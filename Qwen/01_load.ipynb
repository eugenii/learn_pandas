#### Загрузка данных

```python
# Импорты.
import pandas as pd
```

```python
# Загрузка данных из файла *.csv

data = pd.read_csv('kinopoisk-top250.csv')    # читаем данные
print(data.info())                            # выводим информацию о данных 
print(data.describe())
print(data.head())                            # выводим первые 5 строк данных

print(data.columns)                           # выводим названия столбцов

print(data.shape)  # Размер DataFrame (строки, столбцы)
print(data.dtypes)  # Типы данных в каждом столбце
```

```python
# Выборочная загрузка данных

data_1 = pd.read_csv('kinopoisk-top250.csv', usecols=['movie', 'year', 'rating_ball'])
print(data_1.info())

data_2 = pd.read_csv('kinopoisk-top250.csv', usecols=['movie', 'year', 'rating_ball'], nrows=25)
print(data_2.info())

data_skip_rows = pd.read_csv('kinopoisk-top250.csv', usecols=['movie', 'year', 'rating_ball'], skiprows=[1, 2, 3])

# ошибка при загрузке:
data_wrong_sep = pd.read_csv('kinopoisk-top250.csv', sep=';', usecols=['movie', 'year', 'rating_ball']) # wrong
```

```python
# Работа с индексом
df = pd.read_csv("kinopoisk-top250.csv", index_col="movie")  # Указываем столбец с индексом. Можно указать номер столбца (с нуля)

# Вывод первых строк
print(df.head())
print(df.index)
```
